```python
#create engine
import os
from sqlalchemy import create_engine
enginePath = os.path.abspath('stock.sqlite')
engine = create_engine("sqlite+pysqlite:///"+enginePath, echo=True, future=True)

```

```python
#declarative mapping
from sqlalchemy.orm import declarative_base
from sqlalchemy import Column, Integer, String, Float
Base = declarative_base()
class Stock(Base):
    __tablename__ = "stockInTaiwan"
    id = Column(Integer,primary_key=True)
    證券代號 = Column(String(20))
    證券名稱 = Column(String(20))
    成交股數 = Column(Integer)
    成交金額 = Column(Integer)
    開盤價 = Column(Float)
    最高價 = Column(Float)
    最低價 = Column(Float)
    收盤價 = Column(Float)
    漲跌價差 = Column(Float)
    成交筆數 = Column(Integer)
```

```python
from sqlalchemy.orm import Session
session = Session(engine)
stock=session.get(Stock,4)
stock.證券名稱
```

```python
query=session.query(Stock)
query
```

```python
filterObj=query.filter(Stock.成交股數 <= 25000)
stockList=filterObj.all()
for stock in stockList:
    print(stock.證券名稱)
```

```python
for stock in session.query(Stock).filter(Stock.成交股數 <= 25000).all():
    print(stock.證券名稱)
```

```python
for stock in session.query(Stock).filter(Stock.證券名稱 == '台積電').all():
    print('開盤價',stock.開盤價)
    print('收盤價',stock.收盤價)
    print('漲跌價差',stock.漲跌價差)
```

```python
for stock in session.query(Stock).filter(Stock.證券名稱.in_(['台積電', '鴻海', '味全'])).all():
    print('證券名稱',stock.證券名稱)
    print('開盤價',stock.開盤價)
    print('收盤價',stock.收盤價)
    print('漲跌價差',stock.漲跌價差)
    print("===================")
```

```python
for name,code in session.query(Stock.證券名稱, Stock.證券代號):
    print(name)
    print(code)
```
