```python
from sqlalchemy import create_engine
from sqlalchemy import Column,Integer,Float,String
from sqlalchemy.ext.declarative import declarative_base

Base = declarative_base()
class Store(Base):
    __tablename__ = '借問站'
    id = Column(Integer,primary_key=True)
    店家照片 = Column(String,index=True)
    店名_中 = Column(String)
    店名_英 = Column(String)
    店名_日 = Column(String)
    地址_中 = Column(String)
    地址_英 = Column(String)
    地址_日 = Column(String)
    緯度 = Column(Float)
    經度 = Column(Float)
    電話 = Column(String)
    網站 = Column(String)
    營業時間_中 = Column(String)
    營業時間_英 = Column(String)
    營業時間_日 = Column(String)
    店家介紹_中 = Column(String)
    店家介紹_英 = Column(String)
    店家介紹_日 = Column(String)
    交通資訊_中 = Column(String)
    交通資訊_英 = Column(String)
    交通資訊_日 = Column(String)
    店長照片 = Column(String)
    縣市名稱_中 = Column(String)
    縣市名稱_英 = Column(String)
    縣市名稱_日 = Column(String)


conn = create_engine('sqlite:///借問站.sqlite',echo=True)
Base.metadata.create_all(conn)
```

```python
import csv
from sqlalchemy import create_engine
from sqlalchemy.orm import Session
conn = create_engine('sqlite:///借問站.sqlite',echo=True)
session = Session(conn)
with open('借問站.csv',encoding='utf-8') as file:
    next(file)
    rows = csv.reader(file)
    for row in rows:
        store = Store()
        store.店家照片 = row[4]
        store.店名_中 = row[5]
        store.店名_英 = row[6]
        store.店名_日 = row[7]
        store.地址_中 = row[8]
        store.地址_英 = row[9]
        store.地址_日 = row[10]
        store.緯度 = float(row[11])
        store.經度 = float(row[12])
        store.電話 = row[13]
        store.網站 = row[14]
        store.營業時間_中 = row[15]
        store.營業時間_英 = row[16]
        store.營業時間_日 = row[17]
        store.店家介紹_中 = row[18]
        store.店家介紹_英 = row[19]
        store.店家介紹_日 = row[20]
        store.交通資訊_中 = row[21]
        store.交通資訊_英 = row[22]
        store.交通資訊_日 = row[23]
        store.店長照片 = row[32]
        store.縣市名稱_中 = row[35]
        store.縣市名稱_英 = row[36]
        store.縣市名稱_日 = row[37]
        try:
            session.add(store)
        except:
            session.rollback()
            print("出錯")
    session.commit()
```
